# SIR MODEL

SIR Model is a simpler model of spread of a disease  in a closed environment in 
which there are no births, deaths, immigration, or emigration.

The name (SIR) derives from the following three populations considered in the 
model: 

Susceptibles [S] have no immunity from the disease

Infecteds [I] have the disease and can spread it to others 

Recovereds [R] have recovered from the disease and are immune to further infection.
The model gives the differential equation for the rate of change for each of these populations. 

We assume that after a certain amount of time, an individual with the flu recovers. 
Thus, the rate of change of the number of recovereds is proportional to the 
number of infecteds.

## Setting up the Differential Equations for the SIR Model

With the constant of proportionality being the recovery rate (a), differential equation for the rate of change of the number of recovereds can be written as:

dR/dt = aI 

In the SIR model, recovery rate = 1/(number of days infected). For example, if a person is sick with the flu for 2 days, then d = 2 and a = 0.5/day, so that approximately half the infected people get well in a day.

A susceptible person becomes infected by having contact with an infected person. 

The number of such possible contacts is the product of the sizes of the two populations, Susceptibles and Infecteds ,SI.

Rate of change of susceptibles with respect to time: 

dS/dt = –rSI 

for positive constant of proportionality r. 
The constant r, called the transmission constant, reflects the extent and the 
infectiousness of the disease and the interactions among people.

For example if r is 0.002 per day. Thus, 0.002 = 0.2% of the total number of 
possible contacts, SI, results in the disease being spread from one  to another.

With r being a positive number, rate of change of susceptibles is negative. Since this is a closed system, nobody is leaving or entering, number of suspectibles are decreasing, because they are getting sick and then recovered.


Rate of change of Infecteds:
dI/dt = –dS/dt – dR/dt

```python
import matplotlib.pyplot as plt
susceptibles= 762
transmission_constant = 0.00218
infecteds = 1
get_sick = transmission_constant * susceptibles * infecteds
recovery_rate = 0.5
recover = recovery_rate * infecteds
recovereds= 0
t=0
dt=0.25
timeLst=[t]
SLst=[susceptibles]
ILst=[infecteds]
RLst=[recovereds]
while t<14:
    t=t+dt
    susceptibles = susceptibles-get_sick*dt
    recovereds=recovereds+recover*dt
    infecteds = infecteds+(get_sick-recover)*dt
    get_sick = transmission_constant * susceptibles * infecteds
    recover = recovery_rate * infecteds
    timeLst.append(t)
    ILst.append(infecteds)
    SLst.append(susceptibles)
    RLst.append(recovereds)
plt.plot(timeLst,ILst,'g')
plt.text(0.5,50,'Infecteds',color='g')
plt.plot(timeLst,SLst,'r')
plt.text(5.5,650,'Susceptibles',color='r')
plt.plot(timeLst,RLst,'b')
plt.text(5.5,1,'Recovereds',color='b')
plt.title('Graphs of susceptibles (S), infecteds (I), and recovereds (R) versus time (t) in days')
plt.show()
```
